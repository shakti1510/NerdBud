### A Personalized AI-Based Learning Companion

NerdBud adapts programming education based on learner performance,
mistakes, and learning behavior.

This notebook initializes the user interface and captures learner intent.

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import pandas as pd
import os
from datetime import datetime
```

```python
BASE_PATH = os.path.abspath(os.path.join(os.getcwd(), ".."))
LOG_DIR = os.path.join(BASE_PATH, "data", "sample_logs")
LOG_PATH = os.path.join(LOG_DIR, "user_session.csv")

os.makedirs(LOG_DIR, exist_ok=True)
```

```python
username = widgets.Text(
    description="Username:",
    placeholder="Enter your name"
)

language = widgets.Dropdown(
    options=["Python"],
    description="Language:"
)

level = widgets.Dropdown(
    options=["Beginner", "Intermediate", "Advanced"],
    description="Level:"
)

start_button = widgets.Button(
    description="Start Learning",
    button_style="success"
)

display(username, language, level, start_button)
```

```python
def start_session(button):
    clear_output()

    user_data = {
        "username": username.value,
        "language": language.value,
        "level": level.value,
        "timestamp": datetime.now()
    }

    df = pd.DataFrame([user_data])
    df.to_csv(LOG_PATH, mode="a", header=not os.path.exists(LOG_PATH), index=False)

    display(f"Welcome {username.value} ðŸ‘‹")
    display(f"Language: {language.value}")
    display(f"Level: {level.value}")
    display("Session started successfully!")

start_button.on_click(start_session)
```
